### **Что такое множества?**

Если мы постараемся ответить на этот вопрос, то поймем, что:

> **Множество — одно из ключевых понятий математики, представляющее собой набор, совокупность каких-либо объектов — элементов этого множества.**

Видим, что это опеределение пришло к нам из математики. Что не так то и удивительно - многое в программирование пришло оттуда!

Но что самое главное, что мы должны знать?
> Множество - совокупность уникальных объектов! **И это означает, что во множестве присутствуют только уникальные значения - никаких дубликатов.**

С этими знаниями можем идти вперед!

```python
# Создадим множества!

# set() позволяет создать пустое множество. Или же можем сразу наполнить его, передав в качестве аргумента коллекцию.
S_1 = set([1,2,3])

# Или же можем создать множество с помощью фигурных скобок.
S_2  = {1,2,3}

print(S_1)
print(S_2)

# В обоих случаях получаем один и тот же итог.
```

```python
# Теперь посмотрим на работу множеств!
# Проверим действительно ли множество предствляет из себя набор уникальных значений!

S_1 = set([1,2,3,2,3,2,3,2,3])

S_2 = set(['текст','текст',(1,2),'текст',3,'текст',3])

S_3 = set([True, True,(1,2,3),True,2,(1,2,3),'текст'])


print(f'1 Пример: {S_1}')
print(f'2 Пример: {S_2}')
print(f'3 Пример: {S_3}')
```

Увидели что работает! Идем дальше!

**Самое главное, на что мы сейчас обратим внимание, так это то что в функцию `set()` мы можем запихнуть коллекцию!**

А вспоминаем что такое коллекция? А это:
- Списки
- Кортежи
- Строк
- Множества (теперь, после знакомства, и множества)
- Словари (о них позже)

Попробуем использовать в качестве аргумента все по-немногу!

```python
# Проверяем списки

M = [False, 2, 'цифра', 2, 'цифра', 2, 2]
s1 = set(M)

# Проверяем кортежи

K = ('текст', 2, 'текст', 2, True, True)
s2 = set(K)

# Проверяем текст

T = 'молоко'
s3 = set(T)

print(f'Из списка получаем: {s1}')
print(f'Из кортежа получаем: {s2}')
print(f'Из текста получаем: {s3}')
```

Как мы видим - все сработало!
Действительно все коллекции превращаются во множества!

Но на что здесь стоит обратить внимание - значения никак не отсортированы и все не зря! **Множества - неупорядоченная коллекция!**

### **Основные инструменты.**

Давайте рассмотрим большинство возможностей, что предоставляет нам Python!

Так как множество - все та же коллекция, то к ней также применими большинство инструментов, что применимы к остальным коллекциям:

```python
# Создадим для примера множество из слова "Молоко"
s = set('молоко')
print(s)

# Проверить содержится ли элемент во множестве:
print('о' in s)

# Выяснить размер множества:
print(len(s))

# Перебрать элементы:
for item in s:
  print(item, end=' ')
```

**Теперь же рассмотрим инструменты для изменения множеств:**

```python
# Создадим для примера множество.
s = {1, 'python','meow', 2, 'старое'}
print(f'{s} - Создали.')
```

**Добавление:**

```python
# Добавление элемента во множество:
s.add('новое')
print(f'{s} - Добавили')


s.add('новое') # Добавим тот же элемент снова.
print(f'{s} - Добавили снова.') # Увидим, что так как элемент такой уже есть, ничего уже не изменилось.
```

**I Способ удаления:**

```python
# Удаление элемента из множества:

s.remove('старое')
print(f'{s} - Удалили.')
```

```python
# При этом при удалении несуществующего элемента возникает ошибка.

s.remove('несуществующее')
print(f'{s} - Удалили.')
```

**II Способ удаления:**

```python
# Удаление элемента из множества:

s.discard('meow')
print(f'{s} - Удалили.')


# При этом если элемента нет, программа просто ничего не изменяет.
s.discard('несуществующее')
print(f'{s} - Удалили.')
```

Дополнительное:

```python
# Скопировать множество:

s1 = s.copy()
print(f'{s} - Создали Копию.')
```

```python
# Вернуть и удалить произвольный элемент множества

print(f'{s1} - Вывели оригинал.')

x = s1.pop() # Сохраняем удаленный элемент.
print(f'{s1} - После удаления произвольного элемента.')
print(f'{x} - Удаленный элемент.')
```

### **Логические инструменты.**

Продолжим изучение инструментов!

**Представим, что у нас есть два множества:**

- Множество А
- Множество B

Посмотрим, какие логически операции мы можем совершать над ними.

**Ниже - визуальные примеры:**

tg_image_3261382064.jpeg

**Посмотрели глазами, а теперь попробуем написать код.**

Создадим два множества - A и B.

```python
# Создание Множеств

A = {2,3,5,6}
B = {2,4,6,8}
```

```python
# Объединение - Все элементы из обоих множеств.

print(A | B)
```

```python
# Пересечение - Те элементы, что есть в обоих множествах.

print(A & B)
```

```python
# Вычитание - Все элементы одного множества убираем из другого.

print(A - B)
```

```python
# Симметричная Разность - Те элементы, что есть и там, и там, но нет в обоих множествах.

print(A ^ B)
```
