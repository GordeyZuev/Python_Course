## **Что такое функция?**

Давайте первым делом обсудим этот вопрос! У кого-то всплывает в голове математика, ведь в ней есть функции вот такого вида **`f(x) = 2x`**, к примеру!

> Здесь `f(x)` - сама функция, `x` - аргумент функции, а `2x` - то, что выполняет сама функция, а если конкретно, в нашем случае происходит удваивание аргумента.

**В программировании и компьютерах все происходит похожим образом!** У нас есть какая-то функция! И, как правило, это какой-то набор команд! Мы можем производить эти команды как над какими-то данными (или аргументами, как в нашем случае), или же можем просто вызывать эту функцию без аргументов.

С функциями, как бы это странно не показалось, мы сталкивались уже много-много раз. Немного попозже вспомним про это!

*А ниже посмотрим визулизацию работу функции.*

image.png

В данном случае первая фигура слева - какие-то данные, которые мы кладем в нашу функцию (Func), после чего функция **возвращает** нам что-то преобразованное.

Теперь давайте посмотрим как это происходит на примере кода!

```python
# Синтаксис оформления функции.

def func_name(arg1, arg2):
    code code code...
    return ...

# func_name - имя функции.
# code code code...  - тело функции.
# arg1, arg2 - аргументы функции.
```

Насчет **`def`**, что это за странное слово? А это все то же самое сокращение до 3 букв! Происходит `def` от `define`, что в переводе с английского - обозначить. То есть мы просто прописываем что-то на подобии "Обозначить функцию".

> Аргументы функции `(arg1, arg2)` — это её параметры, которые становятся переменными в теле функции.

Как вы видите выше, есть еще очень странная **надпись `return`**. Что это за надпись? Это одна из самых важных строчек в функции - она возвращает итог работы функции - какую-то информацию или данные.

---

*Посмотрели на то, как выглядит функция, теперь давайте напишем какой-нибудь пример. Пусть это будет приветствие человека по имени `name`.*

```python
# Создаем фунцию greeting()

def greeting(name):
  return(f'Hello, {name}! Добро пожаловать!')
```

```python
# Запускаем фунцию для имени Соня
print(greeting("Sonya"))

# Запускаем фунцию для имени Никита
print(greeting("Nikita"))
```

**Видим, что все работает хорошо! Давайте попробуем преобразовать нашу функцию, сделать ее более серьезной!**

```python
# Создаем фунцию greeting()

def new_greeting(name):

  print("Функция выполняется...") # Выводим информацию о запуске функции

  if name.isdigit():
    string = f'Hello, {name.capitalize}! Добро пожаловать!' # Сохраняем приветствие в ситуации верно написанного имени
  else:
    string = 'Добро пожаловать! Мы рады вас видеть!' # Сохраняем приветствие в ситуации НЕверно написанного имени

  return string # Возвращаем приветствие
```

```python
# Запускаем фунцию для имени Соня
guest_name_1 = "Sonya"
print(new_greeting(guest_name_1))

print('-' * 30)

# Запускаем фунцию для имени Андрей
guest_name_2 = "and#rew"
print(new_greeting(guest_name_2))
```

**Важно**, что мы должны заметить выше - аргументом может быть любая переменная. Она необязательно должна называться так, как написано в самой функции - мы можем передавать в качестве аргумента все, а уже в самой функции (для выполнения алгоритма этой самой функции) переменные определяются по новой - только так она сможет работать.

**Также Как мы видим выше, в функцию можно "класть" все - и условия и циклы, и даже другие функции - тот же `print()`.**



Почему это функция, вы спросите? А потому что она также принимает аргументы (то, что мы хотим вывести) и возвразает их (выводит).

## **Особенности.**

Давайте теперь рассмотрим необычные случаи.

- Что будет, если ничего не вернуть?
- Что будет, если не написать аргументы?
- Несколько `return`.

### **Что будет, если не написать аргументы?**

```python
# Обозначаем функцию hello()

def hello():
  return "Hello, world"

print(hello())
print(hello())
```

Как мы видим, да, бывают такие случаи, когда аргументы просто не нужны и это абсолютно нормально!

### **Что будет, если ничего не вернуть?**

```python
# Обозначаем функцию hello_name()

def hello_name(name):
  string = (f"Hello, {name}")

print(hello_name('Андрей'))
```

**Что мы видим здесь? Мы напечатали функцию, но она вернула `None` (То есть, ничего). Причина проста - мы возвращаем ничего, не прописываем `return`.**

В таком случае напрашивается вопрос - "А зачем вообще писать `return`, если можно всегда использовать `print()`". А все просто - вы поймете по примеру ниже.

```python
# Обозначаем функцию name_normilizer_0()

def name_normilizer_0(name):
  name = name.strip(' ,#@') # Очищаем от лишних символов
  name = name.capitalize() # Делаем имя с заглавной буквы

  print(name)

name = ' # гордей ##@ '
new_name = name_normilizer_0(name)
print(f'Вас зовут {new_name}!')
```

**Что мы видим? Мы видим, что никак новое имя мы не сохранили - мы его только вывели! Дальнейшее использование невозможно.**


```python
# Обозначаем функцию name_normilizer_1()

def name_normilizer_1(name):
  name = name.strip(' ,#@') # Очищаем от лишних символов
  name = name.capitalize() # Делаем имя с заглавной буквы

  return name

name = ' # гордей ##@ '
new_name = name_normilizer_1(name)
print(f'Вас зовут {new_name}')
```

**Теперь все отлично! Мы все можем сохранять!**

### **О нескольких return.**

```python
# Обозначим функцию hi()

def hi():
  return 'Hi'
  return 'Hello'

print(hi())
```

**Видим, что функция возвращает только ПЕРВОЕ значение.** Иного быть не может - возврат функции работает только один раз, после этого наша программ просто "Покидает" ее.

То есть если после `return` мы напишем что-то неверное - программа даже не обратит на это внимание, потому что она просто не дойдет до этого фрагмента.

```python
# Обозначим функцию hi_2()

def hi_2():
  return 'Hi'
  return something

print(hi())
```

**Несмотря на то, что у нас нет переменное something - все сработало!**

И исходя из этой особенности мы можем при написании условий не прописывать else! Пример ниже!

```python
# Обозначим функцию even_or_odd()

def even_or_odd(n):
  if n % 2 == 0:
    return 'even'
  return 'odd'
```

А делать, опять-таки, мы можем это потому, что если у нас отработает `if` - выполнится условие (то есть будет выполнен `return`), после чего программа выйдет из функции.

Если же условие выполнено не будет, программа просто вернет иное слово.

## **Немного примеров.**

Посмотрим на применение функций.

```python
# Обозначаем функцию is_prime() для проверки числа на простоту.

def is_prime(n):

  delim_num = 0 # Переменная для хранения количества делителей.

  for i in range(1, n + 1):
    if n % i == 0: # Если число n делится на i без остатка, то i - делитель
      delim_num += 1 # В таком случаем увеличиваем количество делитилей.

  if delim_num == 2: # Если делителя 2 - число простое.
    return True # Вощвращаем True

  return False # Возвращаем False

print(is_prime(17)) # Простое - True
print(is_prime(4)) # Не простое - False
```

### **Итог. Для чего нужны функции?**

Как видно из примера выше, мы можем теперь всего лишь за одну строчку понять, является ли число простым или нет.

И, в общем-то, это основная суть большинства инструментов в программировании - как можно сильнее уменьшить количество повторяемых элементов кода! Ведь если для того, чтобы узнать просто ли число или нет мы тратили 7 строк, для 2 чисел - 14, а для 100 - целых 700, то тут нам достаточно всего-лишь один раз прописать алгоритм и вызывать его сколько хочешь раз, **в разы сократив код**, **сделав его быстро изменяемым** (изменил все в одном месте - работает везде) и **красивым**.
